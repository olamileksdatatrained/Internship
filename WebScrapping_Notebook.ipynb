Question No 2:

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

def scrape_presidents_data(url):
    html = requests.get(url)
    soup = BeautifulSoup(html.content, 'html5lib')

    data = {'Name': [], 'Starting Date': [], 'Closing Date': []}

    table = soup.find('table')
    for row in table.find('tbody').find_all('tr')[2:]:
        columns = row.find_all('td')
        data['Name'].append((columns[0].get_text()))
        data['Starting Date'].append((columns[1].get_text()))
        data['Closing Date'].append((columns[2].get_text()))

    df = pd.DataFrame(data)
    return df

def main():
    url = 'http://currentaffairs.adda247.com/list-of-presidents-of-india/'
    df = scrape_presidents_data(url)
    print(df)

if __name__ == "__main__":
    main()
```

Question No 3, Top 10 ODI Teams:

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

def top_10_teams(url):

    html = requests.get(url)
    soup = BeautifulSoup(html.content, 'html.parser')
    table = soup.find('table', class_='table')

    data = {'TEAM': [], 'MATCHES': [], 'POINTS': [], 'RATING': []}

    for row in table.find('tbody').find_all('tr')[:10]:
        columns = row.find_all('td')
        team = columns[1].text.strip()
        matches = columns[2].text.strip()
        points = columns[3].text.strip()
        rating = columns[4].text.strip()
        
        data['TEAM'].append(team)
        data['MATCHES'].append(matches)
        data['POINTS'].append(points)
        data['RATING'].append(rating)

    df = pd.DataFrame(data, index=range(1,11))
    print(df)

# Starting porint
if __name__ == '__main__':

    url = 'https://www.icc-cricket.com/rankings/mens/team-rankings/odi'

    #Calling function
    top_10_teams(url)
```

Question No 3, Top 10 ODI Players:

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

def top_10_player(url):
    html = requests.get(url)

    data = {'PLAYERS':[], 'TEAMS':[], 'RATINGS':[]}

    soup = BeautifulSoup(html.content, 'html5lib')
    info_div = soup.find('div', {'class': 'rankings-block__banner--player-info'})
    name = info_div.find('div',{'class': 'rankings-block__banner--name'}).get_text()
    team_rating= info_div.find('div', {'class': 'rankings-block__banner--nationality'}).get_text().split()
    team = team_rating[0]
    rating = team_rating[1]
    data['PLAYERS'].append(name)
    data['TEAMS'].append(team)
    data['RATINGS'].append(rating)

    # Extracting the table
    table = soup.find('table', {'class': 'table rankings-card-table'})

    # Extracting Players Info
    for row in table.find('tbody').find_all('tr')[:10]:
        columns = row.find_all('td')
        data['PLAYERS'].append((columns[1].get_text().strip()))
        data['TEAMS'].append((columns[2].get_text().strip()))
        data['RATINGS'].append((columns[3].get_text().strip()))


    # Creating new dataframe
    df = pd.DataFrame(data, index=range(1,11))
    print(df)

# Starting point
if __name__ == '__main__':

    url = 'https://www.icc-cricket.com/rankings/mens/player-rankings/odi'

    # Calling function
    top_10_player(url)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd


def news_details(url):

    html = requests.get(url)
    soup = BeautifulSoup(html.content, 'html5lib')
    list_div = soup.find('ul', {'class': 'LatestNews-list'})

    data = {'Headline': [], 'Time': [], 'News Link': []}

    count = 0
    for row in list_div:
        a_tag = row.find('a')
        link = a_tag['href']
        time = row.find('time')
        title = a_tag.get_text()
        data['News Link'].append(link)
        data['Time'].append(time.get_text())
        data['Headline'].append(title)
    

    df = pd.DataFrame(data)
    print(df)


# Starting point
if __name__ == '__main__':

    url = 'https://www.cnbc.com/world/?region=world'

    # Calling Function
    news_details(url)
```

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup


def ai_articles(url):
   html = requests.get(url)
   soup = BeautifulSoup(html.content, 'html5lib')
   ul = soup.find('ul', {'class': 'sc-9zxyh7-0 cMKaMj'})
   
   data = {'Paper Title': [], 'Authors': [], 'Published Date': [], 'Paper Url': []}

   # Scraping data
   for row in ul:
    a_tag = row.find('a')
    paper_url = a_tag['href']
    paper_title = a_tag.get_text()
    authors = row.find('span', {'class': 'sc-1w3fpd7-0 dnCnAO'}).get_text()
    published_date = row.find('span', {'class': 'sc-1thf9ly-2 dvggWt'}).get_text()

    data['Paper Title'].append(paper_title)
    data['Authors'].append(authors)
    data['Published Date'].append(published_date)
    data['Paper Url'].append(paper_url)

    df = pd.DataFrame(data)
    print(df)

# Starting Point
if __name__ == '__main__':
   
   url = 'https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles'

   # Calling Function
   ai_articles(url)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

def resturant_details(url):

    html = requests.get(url)
    soup = BeautifulSoup(html.content, 'html5lib')

    data = {'Restaurant Name': [], 'Cuisine': [], 'Location': [], 'Ratings': [] ,'Image URL': []}

    image_url = soup.find('img', {'class': 'rdp-banner_img'})['src']
    hotel_name = soup.find('div', {'class': 'restnt-details_info'}).find('h1').get_text()
    cuisine = soup.find('div', {'class': 'about-info d-flex'}).find_all('a')
    location_tags = soup.find('div', {'class': 'restnt-name'}).find_all('a')
    location = location_tags[0].get_text() + location_tags[1].get_text()
    rating = soup.find('div', {'class': 'rest-rating rating-4_5'}).get_text()

    data['Cuisine'].append(cuisine[1].get_text())
    data['Ratings'].append(rating)
    data['Location'].append(location)
    data['Restaurant Name'].append(hotel_name)
    data['Image URL'].append(image_url)

    df = pd.DataFrame(data)
    print(df)


# Starting Point
if __name__ == '__main__':

    url = 'https://www.dineout.co.in/mumbai/kake-da-hotel-vikhroli-west-central-suburbs-89882'

    # Function Calling
    resturant_details(url)
```
