```python
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys

chromedriver_path = "./drivers/chromedriver"

chrome_options = Options()
chrome_options.add_argument("--headless=new")

service = Service(executable_path=chromedriver_path)

driver = webdriver.Chrome(service=service, options=chrome_options)

driver.get('https://www.flipkart.com/')

data = {"Brand Name":[], "Smartphone Name":[], "Colour":[], "RAM":[], "Storage(ROM)":[], "Primary Camera":[],
"Secondary Camera":[], "Display Size":[], "Battery Capacity":[], "Price":[], 'Product Url':[]}

search_bar = driver.find_element(By.NAME, 'q')

search_bar.clear()
search_bar.send_keys('Oneplus Nord')
search_bar.send_keys(Keys.ENTER)
time.sleep(10)

main_div = driver.find_elements(By.XPATH, "//div/div[@class='_1AtVbE col-12-12']")

for div in main_div:
    try:
        link=div.find_element(By.XPATH, ".//div/div/div/a[@class='_1fQZEK']").get_attribute('href')
        body = div.find_element(By.XPATH, ".//div/div/div/a/div[@class='_3pLy-c row']").text.splitlines()
        #print(body)
        data["Brand Name"].append("OnePlus")
        data["Smartphone Name"].append(body[0].split('(')[0])
        data["Colour"].append(body[0].split('(')[1].split(',')[0])
        data['RAM'].append(body[2].split('|')[0])
        data['Storage(ROM)'].append(body[2].split('|')[1])
        data['Display Size'].append(body[3])
        try:
            data['Primary Camera'].append(body[4].split('|')[0])
            data['Secondary Camera'].append(body[4].split('|')[1])
        except:
            data['Secondary Camera'].append('-')

        data['Battery Capacity'].append(body[5])
        data['Price'].append(body[7])
        data['Product Url'].append(link)
    except:
        continue

df = pd.DataFrame(data, index=range(1,25))
df.to_csv('Phone_Details.csv')
```

```python
import time
import pandas as pd
from bs4 import BeautifulSoup
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys

chromedriver_path = "./drivers/chromedriver"

chrome_options = Options()
chrome_options.add_argument("--headless=new")

service = Service(executable_path=chromedriver_path)

driver = webdriver.Chrome(service=service, options=chrome_options)

driver.get('https://www.youtube.com/')


wait = WebDriverWait(driver, 3)
presence = EC.presence_of_element_located
visible = EC.visibility_of_element_located

search_bar = driver.find_element(By.NAME, 'search_query')

search_bar.clear()
search_bar.send_keys('Jawan trailer')
search_bar.send_keys(Keys.ENTER)
time.sleep(3)

WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, "//h3[@class='title-and-badge style-scope ytd-video-renderer']"))).click()

driver.maximize_window()
time.sleep(4)
driver.execute_script("window.scrollTo(0, 500);")


for i in range(0, 45):
    time.sleep(7)
    driver.execute_script("window.scrollTo(0, 100000);")


time.sleep(30)
comments_section = driver.find_element(By.XPATH, '//ytd-comments[@id="comments"]').text

data_list = comments_section.splitlines()

# Initialize empty lists for comments, time, and upvotes
comments = []
times = []
upvotes = []

# Initialize variables to hold temporary data
current_time = None
current_comment = None

# Iterate through the data list
for item in data_list:
    if item.endswith(' ago') and item.count(' ') <= 4:
        # Extract time
        current_time = item if item else "-"
    elif item.isdigit():
        # Extract upvotes
        upvotes.append(item if item else "-")
        # Append the extracted data to respective lists
        if current_comment is not None:
            comments.append(current_comment if current_comment else "-")
            times.append(current_time if current_time else "-")
        # Reset current_comment and current_time
        current_comment = None
        current_time = None
    elif current_time is not None and not item.startswith('@'):
        # This is a comment and not starting with "@"
        if current_comment is None:
            current_comment = item if item else "-"
        else:
            current_comment += " " + item if item else "-"


# Create a dictionary of lists
data = {'Comment': comments, 'Time': times, 'Upvote': upvotes}

# Print the result
for i in range(len(data['Comment'])):
    data['Comment'][i]
    data['Time'][i]
    data['Upvote'][i]


length_index = len(data['Comment'])
data['Upvote'] = data['Upvote'][:length_index]
df=pd.DataFrame(data)
df.to_csv('Youtube_comments.csv',index=False)


driver.close()
```

```python
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys

chromedriver_path = "./drivers/chromedriver"

chrome_options = Options()
chrome_options.add_argument("--headless=new")

service = Service(executable_path=chromedriver_path)

driver = webdriver.Chrome(service=service, options=chrome_options)

data = {
    "Hostel name": [],
    "Distance from city centre": [],
    "Total Reviews": [],
    "Reviews": [],
    "Facilities": []
}

driver.get('https://www.hostelworld.com/pwa/wds/s?q=London,%20England&country=London&city=London&type=city&id=3&from=2023-10-01&to=2023-10-04&guests=2&page=1')

divs = driver.find_elements(By.CLASS_NAME, 'property-card')

current_entry = {} 

for div in divs:
    try:
        lines = div.text.split('\n')  
        if len(lines) > 0:
            hostel_name = lines[0]
            current_entry["Hostel name"] = hostel_name

            for line in lines[1:]:
                if "km from city centre" in line:
                    current_entry["Distance from city centre"] = line
                elif "Total Reviews" in line:
                    current_entry["Total Reviews"] = line
                elif line.startswith("Reviews"):
                    current_entry["Reviews"] = line
                elif line.startswith("Facilities"):
                    current_entry["Facilities"] = line

        for key in data:
            data[key].append(current_entry.get(key, "-"))
    except Exception as e:
        continue

df=pd.DataFrame(data)
df.to_csv('Hostels.csv',index=False)
```
